# 라이브러리 임포트

```python
import boto3
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tqdm import tqdm
from tensorflow.keras.applications import EfficientNetB4
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
import concurrent.futures
import random
import seaborn as sns
```

# 경로 및 하이퍼 파라미터 설정

```python
# AWS S3 버킷 이름
s3_bucket_name = 'dataset2023'
# AWS S3 클라이언트 생성
s3_client = boto3.client('s3')
# S3 버킷의 객체를 리스트하는 Paginator 생성
paginator = s3_client.get_paginator('list_objects_v2')

# 이미지 크기 설정
image_size = 380

# 이미지 라벨 (original 또는 fake)
labels = ('original', 'fake')

# 데이터셋을 가져올 디렉토리와 미니배치 크기를 설정합니다.
data_dir = "v1"
batch_size = 12

# 총 샘플 수를 정의합니다.
#num_samples = 300220
num_samples = 50000
#num_samples_per_label = [128692,171528]
num_samples_per_label = [25000,25000]

# 데이터셋 크기를 계산합니다. (배치 수)
#dataset_size = num_samples // batch_size
#print("Dataset size:", dataset_size)

# 데이터 분할 비율을 지정합니다.
train_size = int(0.8 * num_samples)
val_size = num_samples - train_size

epochs = 15
```

# 데이터 로드 및 제너레이터, 증강 함수 정의

```python
# 이미지 로드 함수를 정의합니다.
def load_image(image_key):
    try:
        # S3로부터 이미지 객체를 가져옵니다.
        response = s3_client.get_object(Bucket=s3_bucket_name, Key=image_key)
        # 이미지를 바이트 배열로 읽어서 numpy 배열로 변환합니다.
        img_bytes = response['Body'].read()
        img_arr = cv2.imdecode(np.asarray(bytearray(img_bytes), dtype=np.uint8), cv2.IMREAD_COLOR)
        # 이미지를 RGB 형식으로 변환합니다. (필요한 경우)
        img_arr = cv2.cvtColor(img_arr, cv2.COLOR_BGR2RGB)
        return img_arr
    except Exception as e:
        raise Exception(f"Error loading image {image_key}: {e}")


# 데이터를 로드하는 제너레이터 함수를 정의합니다.
def data_generator(data_dir, num_samples, is_training=True):
    def generator():
        # 전체 데이터에 대한 상태바를 설정합니다.

        for label, num_samples in zip(labels, num_samples_per_label):
            s3_prefix = f"{data_dir}/{label}"
            response = paginator.paginate(Bucket=s3_bucket_name, Prefix=s3_prefix)

            image_keys = []
            for page in response:
                for obj in page['Contents']:
                    if not obj['Key'].endswith('/'):
                        image_keys.append(obj['Key'])
                        
            random.seed(42) 
            random.shuffle(image_keys)  # 이미지 키를 무작위로 섞습니다.

            image_keys = image_keys[:num_samples]

            for image_key in image_keys:
                img_arr = load_image(image_key)
                # 학습 중에만 데이터 증강을 적용합니다.
                if is_training and np.random.rand() < 0.4:
                    img_arr = augment_image(img_arr)
                
                yield img_arr, labels.index(label)
            

    return generator

# 실시간 데이터 증강 적용을 위한 함수를 정의합니다.
def augment_image(img_arr):
    # Random rotation up to 20 degrees
    rotation_angle = np.random.uniform(-20, 20)
    img_arr = rotate_image(img_arr, rotation_angle)

    # Random horizontal flipping
    if np.random.rand() < 0.5:
        img_arr = flip_image(img_arr)

    return img_arr

# 이미지 변환 함수를 정의합니다.
def rotate_image(img_arr, angle):
    rows, cols, _ = img_arr.shape
    rotation_matrix = cv2.getRotationMatrix2D((cols / 2, rows / 2), angle, 1)
    img_arr = cv2.warpAffine(img_arr, rotation_matrix, (cols, rows))
    return img_arr

def flip_image(img_arr):
    img_arr = np.fliplr(img_arr)
    return img_arr
```

# GPU 메모리 성장 설정

```python
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    for gpu in gpus:
        tf.config.experimental.set_memory_growth(gpu, True)
```

# 제너레이터 및 모델 생성 & 훈련

```python
# 데이터 제너레이터 생성 및 데이터셋 준비
train_data_gen = data_generator(data_dir, train_size, is_training=True)
val_data_gen = data_generator(data_dir, val_size, is_training=False)

train_dataset = tf.data.Dataset.from_generator(train_data_gen, output_signature=(
    tf.TensorSpec(shape=(image_size, image_size, 3), dtype=tf.uint8),
    tf.TensorSpec(shape=(), dtype=tf.int32)
))

val_dataset = tf.data.Dataset.from_generator(val_data_gen, output_signature=(
    tf.TensorSpec(shape=(image_size, image_size, 3), dtype=tf.uint8),
    tf.TensorSpec(shape=(), dtype=tf.int32)
))

train_dataset = train_dataset.shuffle(buffer_size=train_size).cache().batch(batch_size).prefetch(buffer_size=tf.data.AUTOTUNE)
val_dataset = val_dataset.shuffle(buffer_size=val_size).cache().batch(batch_size).prefetch(buffer_size=tf.data.AUTOTUNE)

# 데이터 포맷 설정
tf.keras.backend.set_image_data_format('channels_last')

# MirroredStrategy를 생성합니다.
strategy = tf.distribute.MirroredStrategy()

# MirroredStrategy 내에서 모델을 정의합니다.
with strategy.scope():
    base_model = EfficientNetB4(include_top=False, weights=None, input_shape=(380, 380, 3))
    base_model.trainable = True
    inputs = tf.keras.layers.Input(shape=(image_size, image_size, 3))
    x = base_model(inputs, training=True)
    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    x = tf.keras.layers.Dropout(0.3)(x)
    outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)
    model = tf.keras.Model(inputs, outputs)

    optimizer = tf.keras.optimizers.SGD(
        learning_rate=0.0005,
        momentum=0.9,
        nesterov=True
    )

    loss_fn = tf.keras.losses.BinaryCrossentropy()

    model.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])

# 모델 체크포인트 및 조기 종료 콜백 설정
checkpoint_path = "model_checkpoint.h5"

checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    checkpoint_path, 
    monitor='val_loss',
    save_best_only=True,
    mode='min',
    verbose=1
)

early_stopping_callback = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True,
    verbose=1
)

# 모델 훈련
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=epochs,
    callbacks=[checkpoint_callback, early_stopping_callback]
)

# 훈련된 모델을 TensorFlow SavedModel 형식으로 저장
best_model = tf.keras.models.load_model(checkpoint_path)

best_model.save("trained_model", save_format="tf")
```

```python
# Plotting the training and validation loss over epochs
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation Loss')
plt.show()

# Plotting the training and validation accuracy over epochs
plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy')
plt.show()
```

# 혼동행렬

```python
import seaborn as sns
y_pred = best_model.predict(val_dataset)
y_pred = (y_pred > 0.5).astype(int)  # Convert probabilities to binary predictions (0 or 1)
y_true = np.concatenate([y for _, y in val_dataset], axis=0)

# Create the confusion matrix
confusion_matrix = tf.math.confusion_matrix(y_true, y_pred)

# Plot the confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report


# 예측 결과를 이진 클래스로 변환 (예: 0.5 이상인 경우 1로, 그렇지 않으면 0으로)
val_predictions_binary = (y_pred >= 0.5).astype(int)

# 검증 데이터의 실제 레이블을 가져옵니다.
val_labels = np.concatenate([y for x, y in val_dataset])

# classification_report를 사용하여 분류 성능을 출력합니다.
report = classification_report(val_labels, val_predictions_binary)
print(report)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# 검증 데이터의 실제 레이블을 가져옵니다.
val_labels = np.concatenate([y for x, y in val_dataset])

# ROC 곡선을 계산합니다.
fpr, tpr, _ = roc_curve(val_labels, y_pred)

# AUC (ROC Curve 아래 면적)를 계산합니다.
roc_auc = auc(fpr, tpr)

# ROC 커브를 그립니다.
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
